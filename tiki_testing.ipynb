Thinking through how to get recipes I am looking for.

Example: I have rum, lime, simple syrup, and lemon juice. With n ingredients in my pantry, I want it to search for:
* All that require all n ingredients
* All permutations that match n - 1 (while n >=3)
* All permutations that match n - 2 (while

```python
# Baby practice set
pantry = ['rum', 'lime', 'simple syrup']

d = {'id': ['1','2','3'],
     'drink': ['daqauri', 'gimlet', 'sour'],
     'in_total': [3,3,3],
     'rum': [1, 0, 0],
     'lime': [1, 1, 0],
     'simple syrup':[1, 1, 1],
     'lemon': [0,0,1],
     'gin': [0,1,1],
    }

bartender = pd.DataFrame(data=d)
```

```python
found = []

# Creating one list with ids for all items that match any one ingredient
for item in pantry:
    found.extend(bartender[bartender[item] == 1]['id'].tolist())

# Turning that list into a frequency dictionary
freq = {item: found.count(item) for item in found}

# Creating dictionaries for each drink with its frequency count, common for matching all
common_dict = {id: tot for id, tot in zip(bartender.id.tolist(), bartender.in_total.tolist())}
# Close for matching n - 1
close_dict = {id: tot - 1 for id, tot in zip(bartender.id.tolist(), bartender.in_total.tolist())}

# Finding matching items (key, value)
common_id = [id[0] for id in list(freq.items() & common_dict.items())]
# Finding clost matching items (key, value - 1)
almost_id = [id[0] for id in list(freq.items() & close_dict.items())]

#Getting drink name for matches
common_name = bartender[bartender.id.isin(common_id)]['drink'].values
almost_name = bartender[bartender.id.isin(almost_id)]['drink'].values

print("Can make:")
for f in common_name:
    print(f)
print("----")
print("Almost:")
for f in almost_name:
    print(f)
```
